```julia
using PyPlot
```

```julia
x = range(0, 2π, 256)
y = range(-π/2, π/2, 128)
print()
```

```julia
xx = reshape(ones(128)' .* x, (256,128,1))
yy = reshape(y' .* ones(256), (256,128,1))
println()
```

```julia
println(size(xx))
println(size(yy))
```

```julia
# zeta_anomaly = 1e-7 * sin(15*?)*exp(-((?-?)/?)^2)
zeta_anomaly = 1e-7 * sin.(15*xx) .* exp.(-((yy.-(20*π/180))/(5*π/180)).^2)
PyPlot.figure(figsize = (16, 9), dpi = 160)
PyPlot.contourf(xx[:, :, 1]*180/π, yy[:, :, 1]*180/π, zeta_anomaly[:, :, 1], levels = 32, cmap = "bwr")
```

```julia
zeta_basic = 4e-5 .* exp.(-((yy.-(19*π/180))/(3*π/180)).^2)
PyPlot.figure(figsize = (16, 9), dpi = 160)
PyPlot.contourf(xx[:, :, 1]*180/π, yy[:, :, 1]*180/π, zeta_basic[:, :, 1], levels = 32, cmap = "bwr")
```
