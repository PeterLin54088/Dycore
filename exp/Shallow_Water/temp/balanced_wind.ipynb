```julia
using PyPlot
using LinearAlgebra
```

```julia
rad_to_deg = 180/π
deg_to_rad = π/180
@show
```

```julia
Nx = 256
Ny = 128
Nz = 1
x = range(0, 2π, Nx)
y = range(-π/2, π/2, Ny)
xx = reshape(ones(Ny)' .* x, (Nx,Ny,Nz))
yy = reshape(y' .* ones(Nx), (Nx,Ny,Nz))
println("x - coord")
println(size(xx))
PyPlot.figure(figsize = (16, 9), dpi = 160)
PyPlot.contourf(xx[:,:,1], levels = Nx, cmap = "jet")
PyPlot.show()
println("y - coord")
println(size(yy))
PyPlot.figure(figsize = (16, 9), dpi = 160)
PyPlot.contourf(yy[:,:,1], levels = Nx, cmap = "jet")
PyPlot.show()
```

```julia
zeta_amp = 4e-5
zeta_lat = 19*deg_to_rad
zeta_radius = 10*deg_to_rad
zeta_basic = zeta_amp * exp.(-((yy.-zeta_lat)/zeta_radius).^2)
PyPlot.figure(figsize = (16, 9), dpi = 160)
PyPlot.contourf(xx[:, :, 1]*rad_to_deg, yy[:, :, 1]*rad_to_deg, 
                zeta_basic[:, :, 1], levels = 128, cmap = "bwr")
```

```julia
function Differential_Matrix(size::Int64)
    instantiate_matrix = 1.0* Matrix(I, size, size)
    matrix = (circshift(instantiate_matrix, (0,1)) - circshift(instantiate_matrix, (0,-1))) ./ 2.0
    matrix[1,:] .= 0
    matrix[size,:] .= 0
    matrix[1,1], matrix[1,2] = -1, 1
    matrix[size,size-1], matrix[size,size] = -1, 1
    matrix 
    return matrix
end
```

```julia
D = Differential_Matrix(10)
```
